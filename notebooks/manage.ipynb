```python
cols = ['loan_amnt', 'term', 'int_rate', 'installment', 'emp_length', 'annual_inc', 'fico_mean', 'credit_history_years', 'dti', 'inq_last_6mths', 'delinq_2yrs', 'acc_now_delinq', 'open_acc', 'total_acc', 'revol_bal', 'revol_util', 'bc_util', 'percent_bc_gt_75', 'total_rev_hi_lim', 'avg_cur_bal', 'pub_rec', 'pub_rec_bankruptcies', 'tax_liens', 'collections_12_mths_ex_med', 'mort_acc', 'pct_tl_nvr_dlq', 'num_actv_bc_tl', 'num_actv_rev_tl', 'num_op_rev_tl', 'num_rev_tl_bal_gt_0', 'num_tl_op_past_12m']

for col in cols:
    print("Unique values:" + str(df[col].value_counts()), end='\n\n')
    print("Empty values:" + str(df[col].isna().sum()), end='\n\n')
```

```python
df['inq_last_6mths'] = df['inq_last_6mths'].clip(upper=6)
df['delinq_2yrs'] = df['delinq_2yrs'].clip(upper=10)
df['acc_now_delinq'] = df['acc_now_delinq'].clip(upper=1)
df['collections_12_mths_ex_med'] = df['collections_12_mths_ex_med'].clip(upper=3)
df['pub_rec'] = df['pub_rec'].clip(upper=5)
df['pub_rec_bankruptcies'] = df['pub_rec_bankruptcies'].clip(upper=3)
df['tax_liens'] = df['tax_liens'].clip(upper=3)
```

```python
df['revol_bal_log'] = np.log1p(df['revol_bal'])
df['total_rev_hi_lim_log'] = np.log1p(df['total_rev_hi_lim'])
df['avg_cur_bal_log'] = np.log1p(df['avg_cur_bal'])

df = df.drop(columns=['revol_bal', 'total_rev_hi_lim', 'avg_cur_bal'], axis=1)
```

```python
df['revol_util_missing'] = df['revol_util'].isna().astype(int)
df['bc_util_missing'] = df['bc_util'].isna().astype(int)
df['percent_bc_gt_75_missing'] = df['percent_bc_gt_75'].isna().astype(int)

df['revol_util'] = df['revol_util'].fillna(0).clip(upper=100)
df['bc_util'] = df['bc_util'].fillna(0).clip(upper=100)
df['percent_bc_gt_75'] = df['percent_bc_gt_75'].fillna(0)
```

```python
# df['term'] = df['term'].map({'36 months': 36, '60 months': 60})
grade_map = {'A':1,'B':2,'C':3,'D':4,'E':5,'F':6,'G':7}
df['grade'] = df['grade'].map(grade_map)
```

```python
df['term'] = df['term'].astype(str).str.extract(r'(\d+)').astype(int)

df['term'].value_counts()
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
df['inq_last_6mths'] = df['inq_last_6mths'].fillna(0)
df['tax_liens'] = df['tax_liens'].fillna(0)
df['collections_12_mths_ex_med'] = df['collections_12_mths_ex_med'].fillna(0)
df['pub_rec_bankruptcies'] = df['pub_rec_bankruptcies'].fillna(0)
```

```python
dti_median = df['dti'].median()
df['dti'] = df['dti'].fillna(dti_median)
```

```python
def covert_emp_length(emp_length):
    if pd.isna(emp_length):
        return np.nan
    emp_length = emp_length.strip()
    if emp_length == '< 1 year':
        return 0
    elif emp_length == '10+ years':
        return 10
    else:
        return int(emp_length.split(' ')[0])

df['emp_length'] = df['emp_length'].apply(covert_emp_length)
emp_length_median = df['emp_length'].median()
df['emp_length'] = df['emp_length'].fillna(emp_length_median)
```

```python
df['mths_since_last_delinq'] = df['mths_since_last_delinq'].fillna(999)
df['mths_since_last_record'] = df['mths_since_last_record'].fillna(999)
```

```python
import xgboost as xgb
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
```

```python
df = df.sort_values(by='issue_d')
```

```python
split_idx = int(len(df) * 0.8)

train_df = df.iloc[:split_idx]
test_df = df.iloc[split_idx:]
```

```python
X_train = train_df.drop(columns=['loan_status', 'issue_d'])
y_train = train_df['loan_status']

X_test = test_df.drop(columns=['loan_status', 'issue_d'])
y_test = test_df['loan_status']
```

```python
log_model = LogisticRegression(penalty='l2', C=1.0, class_weight='balanced', solver='lbfgs', max_iter=1000, random_state=42, n_jobs=-1)

log_model.fit(X_train, y_train)
```

```python
y_pred_proba = log_model.predict_proba(X_test)[:, 1]
roc_auc_log = roc_auc_score(y_test, y_pred_proba)

roc_auc_log
```

```python
scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()
```

```python
xgb_model = xgb.XGBClassifier(n_estimators=300, max_depth=4, learning_rate=0.05, min_child_weight=50, scale_pos_weight=scale_pos_weight, subsample=0.8, colsample_bytree=0.8, random_state=42, n_jobs=-1)
xgb_model.fit(X_train, y_train)
```

```python
y_pred_proba_xgb = xgb_model.predict_proba(X_test)[:, 1]
roc_auc_xgb = roc_auc_score(y_test, y_pred_proba_xgb)

roc_auc_xgb
```
